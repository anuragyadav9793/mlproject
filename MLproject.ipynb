```python
x = iris.drop(['species'], axis=1)
```

```python
x
```

```python
y = iris['species']
```

```python
y
```

```python
le = LabelEncoder()
y = le.fit_transform(y)
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2)
```

add random_state to train_test_split

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state=1)
```

logistic regression

```python
lr = LogisticRegression(solver = 'newton-cg')
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred1 = lr.predict(x_test)
```

get the confusion matrix

```python
confusion_matrix(y_test, y_pred1)
```

```python
sns.heatmap(data = confusion_matrix(y_test, y_pred1),annot=True, cmap = 'Blues')
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
```

get the accuracy score

```python
accuracy_score(y_test, y_pred1)
```

decision tree

```python
dtree = DecisionTreeClassifier()
dtree.fit(x_train, y_train)
```

```python
y_pred2 = dtree.predict(x_test)
```

get the confusion matrix, heatmaap for confusion matrix and accuracy score

```python
confusion_matrix(y_test, y_pred2)
```

```python
sns.heatmap(data = confusion_matrix(y_test, y_pred2),annot=True, cmap = 'Blues')
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
```

```python
accuracy_score(y_test, y_pred2)
```

```python
plt.figure(figsize=(20,20))
dec_tree = plot_tree(decision_tree = dtree, feature_names = iris.columns[:-1],
                     class_names= ("setosa", "vercicolor","verginica"), filled = True)
```
